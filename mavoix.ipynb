The given dataset doesnt have target variable,so this comes under unsupervised learning.Clustering is unsupervised learning i choose that to segment data in 2 groups for Datascience and Web development.
Steps involved:
1.Importing data
2.data cleaning and data manupalation
3.Cluster model

Data cleaning and manupalation:
1.Remove the blank column initially.
2.Degree column is categorical variable as clustering uses only numerical so i convert that into numerical by mentioning different numbers for every unique entry.
3.perfomance_ug,perfomance_12,perfomance_10(i didnt consider perfomance_pg because more than 70% rows are na in that column) are in raw form so split numerator and denominator fetch them into different column(score_ug,ugoutof),but problem is few are in form of cgpa and remaining in form of percentage so i equate outof(12outof,10outof) column into 100 to generalised for all rows and calculate percentage.where ever marks is in form of cgpa i multiplied that with 10 so that i can convert that into percentage.
4.i got 10,12,ug percentage seperatly to fill missing values in these three column take mean of other two columns's rows and insert into row where na is.
5.To do clustering i considered all skills columns,Degree column,year of passing and derived percentage columns.
6.i scale the all considered columns as first step of clustering.
7.Used k-mean clustering, choose n_cluster = 2.I am interested in only only two groups so
8.i got data segmented in form of zeros and ones.data did not segmented 100% accurately however i got moderate accuracy
i assume 1 for Data science,0 for Web development.
9.to validate clustering i used davies bouldin index got score 2.36 lower the score better the model it is.
10.To find closest points in clusters i used pairwise_distances_argmin_min and got index of points which are near to each cluster centroid.

```python
#considering only skills and degree and percentage column for modelling
data_num=data1[['Python (out of 3)','R Programming (out of 3)','Deep Learning (out of 3)','PHP (out of 3)','MySQL (out of 3)','HTML (out of 3)','CSS (out of 3)','JavaScript (out of 3)','AJAX (out of 3)','Bootstrap (out of 3)','MongoDB (out of 3)','Node.js (out of 3)','ReactJS (out of 3)','Current Year Of Graduation','Degree','ugper','12per','10per']]
```

```python
data_num.isna().sum()
```

```python
#scaling all variables
import sklearn.preprocessing as preprocessing
dat_scaled=preprocessing.scale(data_num,axis=0)
```

```python
#loading cluster library
import sklearn.cluster as cluster
```

```python
#k mean clustering, no of clusters 2
kmeans=cluster.KMeans(n_clusters=2,init="k-means++")
kmeans=kmeans.fit(dat_scaled)
```

```python
#segmented in two groups 
kmeans.labels_
```

```python
#cluster centers
j=kmeans.cluster_centers_
```

```python
j
```

```python
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
```

```python
labels=kmeans.labels_
```

```python
#importing davies bouldin score library
from sklearn.metrics import davies_bouldin_score 
```

```python
#davies bouldin score
print(davies_bouldin_score(dat_scaled,labels)) 
```

```python
from sklearn.metrics import pairwise_distances_argmin_min

```

```python
#closest points to each cluster centroid
closest, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, dat_scaled)
closest
```
